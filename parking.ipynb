```python
li = []
li_index = []

for index, illegal_coor in seoul_illegal_parking.iterrows():
    # print(illegal_coor)
    a1 = illegal_coor.loc['경도']
    b1 = illegal_coor.loc['위도']
    count = 0
    length_fix = 10000
    for index_, Re_coor in integrated_df.iterrows():
        a2 = Re_coor.loc['경도']
        b2 = Re_coor.loc['위도']
        length = 133330*math.sqrt(math.cos(37.5)*((a2-a1)**2) + (b2-b1)**2)
        if length_fix > length:
            check = count
            length_fix = length
        count += 1
    if length_fix < 50:
        li.append(50)
    else:
        li.append(length_fix)

    li_index.append(check)

# print(li)
# print(li_index)
```
